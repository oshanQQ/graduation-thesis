```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression


def Plot_as_3d(filename):
    df = pd.read_csv(filename)

    # Split the DataFrame into features (X) and target (y)
    X = df[['batch_size', 'epoch']]
    y = df['learning_time']

    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')

    ax.scatter(df['batch_size'], df['epoch'], df['learning_time'], c='r', marker='o')
    ax.set_xlabel('Batch Size')
    ax.set_ylabel('Epoch')
    ax.set_zlim(0, 1500)
    ax.set_zlabel('Learning Time')

    plt.show()


def Training(filename):
    # Load the CSV file as a pandas DataFrame
    df = pd.read_csv(filename)
    
    # Split the DataFrame into features (X) and target (y)
    X = df[['batch_size', 'epoch']]
    y = df['learning_time']

    # Split the data into training and test sets
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

    # Train the linear regression model
    model = LinearRegression()
    model.fit(X_train, y_train)

    # Evaluate the model on the test data
    test_score = model.score(X_test, y_test)
    print("Test Score:", test_score)


Plot_as_3d('learning_time_on_local.csv')
Training('learning_time_on_local.csv')

Plot_as_3d('learning_time_on_jupyterhub.csv')
Training('learning_time_on_jupyterhub.csv')
```
