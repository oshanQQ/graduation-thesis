```python
import matplotlib.pyplot as plt
import numpy as np
from keras.datasets import mnist, cifar10

# Load MNIST data
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# Plot first 10 images from MNIST
for i in range(10):
    plt.subplot(2, 5, i + 1)
    plt.imshow(x_train[i], cmap='gray')
    plt.axis('off')

plt.suptitle("10 MNIST Images", fontsize=20)
plt.show()

# Load CIFAR-10 data
(x_train, y_train), (x_test, y_test) = cifar10.load_data()

# Plot first 10 images from CIFAR-10
for i in range(10):
    plt.subplot(2, 5, i + 1)
    plt.imshow(x_train[i])
    plt.axis('off')

plt.suptitle("10 CIFAR-10 Images", fontsize=20)
plt.show()
```
